* Mean
    * Arithematic Mean
    $$
    \text{Arithematic Mean or } \bar{x} = \frac{\sum{fx}}{\sum{f}}
    
    \\[15pt]

    \text{Arithematic Mean or } \bar{x} = \text{provisional mean + } \frac{\sum{fD}}{\sum{f}}
    
    \\[5pt]
    $$

    $
    \text{Where }
    \\[5pt]
    \text{D = size of mean - provisional mean}
    $

    * Geometric Mean

    $$
    \text{G.M. of } x = \sqrt[n]{x_1^{f_1} \times x_2^{f_2} \times \dots \times x_n^{f_n}}
    $$

    * Harmonic Mean

    $$\text{H.M = } \frac{\sum{f}}{\sum{f \times \frac{1}{x}}}$$

* Median
    $$
    \text{Median for grouped data = } l1 + \left(\frac{\frac{N}{2} - C.F}{f} \times ( l2 - l1 )\right)
    $$

    $
    \text{Where}
    \\[5pt]
    l1 \Rightarrow \text{lower limit of the median class}
    \\
    l2 \Rightarrow \text{upper limit of the median class}
    \\
    f \Rightarrow \text{frequency of the median class}
    \\
    C.F \Rightarrow \text{cumlative freqency before the median class}
    $

***Example***

Step 1: $\text{Median in the size of }\frac{26600 + 1}{2} = 13300.5 \Rightarrow \text{ Check where the value lies in the C.F}$

Step 2: $\text{Find 13300.5 in the C.F we get between 9.5 - 29.5}$

Step 3: $l1 = 9.5 \quad l2 = 29.5$

Step 3: $f$ = 4560  $\Rightarrow \text{means the value of }l2$
